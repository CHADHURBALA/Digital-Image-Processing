```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

imag = 'tank.png'
img = cv2.imread(imag)
s=img.shape
imgg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
cv2_imshow(imgg)
```

```python
def Histogram(image):
  H= [0]*256
  s = image.shape
  for i in range(s[0]) :
    for j in range(s[1]):
      k=image[i,j]
      H[k] = H[k] + 1

  
  return H
```

```python
cv2_imshow(imgg)
H = Histogram(imgg)
plt.plot(H)
plt.show()
```

```python
# Histogram equalization using inbuilt function
equ = cv2.equalizeHist(imgg)
HE = Histogram(equ)
plt.plot(HE)
plt.show()
cv2_imshow(equ)
```

```python
# Histogram equalization (manual)
y = np.array([])
y = np.append(y,H[0])
for i in range(255):
  k = H[i+1]+y[i]
  y=np.append(y, k)
y = np.round((y/(s[0]*s[1]))*(255))
imgh = imgg
for i in range(s[0]):
  for j in range(s[1]) :
    k = imgg[i, j]
    imgh[i,j] = y[k]
  
cv2_imshow(imgh)
H1 = Histogram(imgh)
plt.plot(H1)
plt.show()
```
